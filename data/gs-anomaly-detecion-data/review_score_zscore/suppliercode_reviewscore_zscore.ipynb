```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.ticker import ScalarFormatter, MultipleLocator
from sklearn.neighbors import LocalOutlierFactor
from scipy.stats import zscore
```

```python
data_route = "C:\\Users\\SSAFY\\workspace\\final_project\\dataset\\dataset_full.csv"
data = pd.read_csv(data_route, encoding='utf-8')
```

```python
data['supplier_review_score_zscore'] = data.groupby('supplier_code')['review_score'].transform(lambda x: zscore(x))
```

```python
# 최소 review_count와 zscore_threshold를 조정하여 보면 됩니다.
review_count = 1
zscore_threshold = -3
```

```python
data[(data['supplier_review_score_zscore'] < zscore_threshold) & 
                            (data['review_count'] >= review_count)
                           ][['prd_id', 'view_name', 'supplier_code', 'review_score', 'review_count', 'supplier_review_score_zscore']].sort_values(by='supplier_code')
```
