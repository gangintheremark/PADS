```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_route = "C:\\workspace\\project_final\\data\\dataset_sample_30percent.csv"
data = pd.read_csv(data_route, encoding='utf-8')
```

```python
def get_discprice_outlier_score(df):
    df['disc_outlier_score'] = (df['discprice'] > df['price'])*100
    return df
```

```python
data = get_discprice_outlier_score(data)
outliers = data[(data['disc_outlier_score'] == 100)]
outliers = outliers.sort_values(by='disc_outlier_score', ascending=False)
```
